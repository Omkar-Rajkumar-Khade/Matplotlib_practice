## Plotting PIE chart

```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
#use explode method to slice pie chart
explode=[0.1,0,0.2,0,0]
plt.pie(class1_students,labels=classes,explode=explode)
plt.show()
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
#add manual color to chart
color=['g','y','m','b','r']
plt.pie(class1_students,labels=classes,colors=color)
plt.show()
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
#increase size of pie chart by radius
color=['g','y','m','b','r']
plt.pie(class1_students,labels=classes,colors=color, radius=1.4)
plt.show()
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
#change start angle manually
plt.pie(class1_students,labels=classes, startangle=90)
plt.show()
```

```python
classes=['Python','R','ML','AI','DS']
class1_students=[45,15,35,25,30]
#add percentage to graph manually by using autopct
plt.pie(class1_students,labels=classes, startangle=90,autopct="%0.1f%%")
plt.show()
```

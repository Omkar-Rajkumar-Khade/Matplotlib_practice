## Plotting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["python","R","AI","ML","DS"]
class1_students=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
#use plt.bar method to plot bar chart
plt.bar(classes,class1_students)
```

```python
#use plt.bah method to plot horizontal bar chart
plt.barh(classes,class1_students)
```

```python
#change width of histogram
plt.bar(classes,class1_students,width=0.3)
```

```python
#change color of histogram & edgewidth
plt.bar(classes,class1_students,width=0.3,color="b",edgecolor="m")
```

```python
#change linestyle of histogram & edgewidth
plt.bar(classes,class1_students,width=0.3,color="b",edgecolor="m",linestyle="--")
```

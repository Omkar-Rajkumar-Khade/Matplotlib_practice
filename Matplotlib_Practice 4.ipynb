## Plotting Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("./Datasets/International_matches.csv")
df
```

```python
x=df["home_team"]
y=df["home_team_fifa_rank"]
```

```python
plt.scatter(x,y)
plt.title("Home team Vs Fifa rank graph")
plt.xlabel("home_team")
plt.ylabel("home_team_fifa_rank")
```

```python
#add color to plt
plt.scatter(x,y,color=
           'green')
plt.title("Home team Vs Fifa rank graph")
plt.xlabel("home_team")
plt.ylabel("home_team_fifa_rank")
```

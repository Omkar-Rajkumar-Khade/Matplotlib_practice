## Plotting Histrograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random 
```

```python
ml_student_age =np.random.randint(18,45,(100))
py_student_age =np.random.randint(15,80,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
#hist method is used add histogram
plt.hist(ml_student_age)
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins=[15,20,25,30,40,45]
#use rwidtth method to add space between bars
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
#use histtype method to change histogram type
plt.hist(ml_student_age,histtype="step")
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
#use orientation method to change histogram orientation
plt.hist(ml_student_age,rwidth=0.6,orientation='horizontal')
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
#use color method to change histogram color
plt.hist(ml_student_age,rwidth=0.6,color='m')
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

### Add 2 graphs in one frame

```python
#method 1
plt.hist(ml_student_age,rwidth=0.6,color='b')
plt.hist(py_student_age,rwidth=0.6,color='y')
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```

```python
#method 2
plt.hist([ml_student_age,py_student_age],rwidth=0.8,color=['b','y'])
plt.title("ML Student age histrogram")
plt.xlabel("Student age category")
plt.ylabel("No. of student age")
plt.show()
```
